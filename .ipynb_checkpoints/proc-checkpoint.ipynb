```python
# Top10
#### This notebook processes the training dataset's text to build classification models.
#### Tests several classifiers with accuracy over dev data.
```

```python
def get_bucket(age):
    if 14 <= age <= 16:
        return 0
    if 24 <= age <= 26:
        return 1
    if 34 <= age <= 36:
        return 2
    if 44 <= age <= 46:
        return 3
    
    if age < 14:
        return 4
    if 16 < age < 24:
        return 5
    if 26 < age < 34:
        return 6
    if 36 < age < 44:
        return 7
    if age > 46:
        return 8
```

```python
import pandas as pd
from collections import defaultdict as dd
filename = "COMP30027_2018S1_proj2-data/train_raw.csv"
train = pd.read_csv(filename, names = ["User ID", "Gender", "Age", "Occupation", "Star Sign", "date", "text"])
```

```python
train
```

```python
filename = "COMP30027_2018S1_proj2-data/dev_raw.csv"
dev = pd.read_csv(filename, names = ["User ID", "Gender", "Age", "Occupation", "Star Sign", "date", "text"])
```

```python
dev
```

```python
# Counts trigrams
buckets = 9
train_trigrams = [dd(int) for i in range(buckets)]
train_url_counts = [dd(int) for i in range(buckets)]

for i in range(len(train)):
    age = train["Age"][i]
    text = train["text"][i].strip()
    bucket = get_bucket(age)
    
    # Counts occurences of urlLink then removes them
    train_url_counts[bucket][text.count("urlLink")] += 1
    text.replace("urlLink ", "").replace(" urlLink", "").replace("urlLink", "")
    
    # Counts trigrams
    for i in range(len(text)-2):
        train_trigrams[bucket][text[i:i+3]] += 1
    
for i in range(buckets):
    print(train_url_counts[i])
    
# Normalises trigram counts
for bucket in range(buckets):
    if not train_trigrams[bucket]:
        print(bucket, "has no contents")
        buckets -= 1 # assumes ommited buckets are at the end of the range
        continue
    max_val = max(train_trigrams[bucket].values())
    for key in train_trigrams[bucket]:
        train_trigrams[bucket][key] = train_trigrams[bucket][key]/max_val
print(buckets)
```

```python
# Classifies all dev instances
classification = []
dev_url_counts = []

# Creates set of all keys
all_keys = set()
for i in range(buckets):
    all_keys.update(train_trigrams[i].keys())

for i in range(len(dev)):
    print(".", end="")
    age = dev["Age"][i]
    text = dev["text"][i].strip()
    bucket = get_bucket(age)
    
    # Counts occurences of urlLink then removes them
    dev_url_counts.append(text.count("urlLink"))
    text.replace("urlLink ", "").replace(" urlLink", "").replace("urlLink", "")
    
    if len(text) < 3:
        print("text is empty:", dev["text"][i])
        classification.append(-1)
        dev_url_counts[-1] = -1
        continue
    
    # Counts trigrams
    dev_trigrams = dd(int)
    for i in range(len(text)-2):
        dev_trigrams[text[i:i+3]] += 1
    
    # Normalising
    max_val = max(dev_trigrams.values())
    for key in dev_trigrams:
        dev_trigrams[key] = dev_trigrams[key]/max_val
    
    all_keys.update(dev_trigrams.keys())
    
    # finding distances between trigrams in this text and aggregated
    distances = [0.0 for i in range(buckets)]
    for key in all_keys:
        for bucket in range(buckets):
            distances[bucket] += abs(train_trigrams[bucket][key] - dev_trigrams[key])
    #print(distances)
    pred = min(distances)
    classification.append(distances.index(pred))
```

```python
correct = 0
for i in range(len(classification)):
    if classification[i] == -1:
        continue
    bucket = get_bucket(dev["Age"][i])
    if bucket == classification[i]:
        correct += 1
print("accuracy: {}%".format(correct/len(classification)*100))
print("Correct:", correct)
print("Total:", len(classification))
```

```python
counts = []
for i in range(buckets):
    count = []
    for j in range(112):
        if j in train_url_counts[i]:
            count.append(train_url_counts[i][j])
        else:
            count.append(0)
    counts.append(count)
counts
```
